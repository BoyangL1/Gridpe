```python
import numpy as np
import matplotlib.pyplot as plt

# === Parameters ===
n = 2  # Dimension (2D)
omega_range = np.linspace(-1, 1, 100)  # Frequency values (1D example)
d_range = np.linspace(-20, 20, 1000)  # Displacement values

# === Define the spectrum c(omega) ===
def c(omega):
    """Spectral envelope (e.g., Gaussian)"""
    return np.exp(-omega**2)  # Since omega is scalar, use scalar norm

# === Compute h(d) ===
h_d = []

for d in d_range:
    sum_term = np.sum(np.abs(c(omega_range))**2 * np.exp(1j * omega_range * d))
    h_val = np.real((2 * np.pi) ** n * sum_term)
    h_d.append(h_val)

# === Plotting ===
plt.figure(figsize=(8, 5))
plt.plot(d_range, h_d, label=r'$h(d)$')
plt.xlabel(r'Relative displacement $d$', fontsize=14)
plt.ylabel(r'$h(d)$', fontsize=14)
plt.title(r'Inner product $h(d) = \langle g(x), g(x + d) \rangle$', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
# Save figure at 300 DPI
plt.savefig("gd.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
